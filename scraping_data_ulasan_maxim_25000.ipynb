```python
# import library

from google_play_scraper import Sort, reviews_all
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
# Mendefinisikan id aplikasi, negara, dan bahasa

app_id = 'com.taxsee.taxsee'  # ID aplikasi Maxim di Google Play Store
country = 'id'  # kode negara Indonesia
language = 'id'  # kode bahasa Indonesia
```

```python
# Scraping ulasan

reviews = reviews_all(
    app_id,
    lang=language,
    country=country,
    sort=Sort.MOST_RELEVANT,  # Mengambil urutan dari yang paling relevant
    count=25000,  # Jumlah maksimum ulasan yang ingin diambil
)
```

```python
# Membuat dataframe dari hasil scraping

maxim_reviews = pd.DataFrame(reviews)
maxim_reviews.head()
```

```python
# Mengambil 25000 data pertama
maxim_reviews_25000 = maxim_reviews.head(25000)
len(maxim_reviews_25000)
```

```python
# Menyimpan data ke file CSV
maxim_reviews_25000.to_csv('ulasan_maxim.csv', index=False)
```
